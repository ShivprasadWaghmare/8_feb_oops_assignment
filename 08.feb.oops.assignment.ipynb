2. Differentiate between Abstraction and Encapsulation. Explain with an example.
ans:
**Abstraction:** Imagine an electric oven. You know you can bake with it, but you don't need to know how the heating elements work. **Abstraction hides the complex details and shows only what's needed.**

3. What is abc module in python? Why is it used?
ans:
 **Here's a comprehensive explanation of the abc module in Python and its uses:**

**What is the abc module?**

- It's a built-in Python module that stands for "Abstract Base Classes."
- It provides the tools to create abstract base classes (ABCs), which are classes that cannot be instantiated directly.
- ABCs serve as blueprints for concrete classes, outlining the methods and properties that subclasses must implement.

**Why is it used?**

- **Enforcing Interfaces:** ABCs ensure that all subclasses adhere to a specific interface, maintaining code consistency and clarity.
- **Defining Contracts:** They establish a contract between classes, guaranteeing that subclasses provide certain functionality.
- **Polymorphism:** They enable polymorphic behavior, meaning code can interact with objects of different subclasses through a common interface.
- **Code Organization:** ABCs promote code organization and reusability by defining common patterns and structures.
- **Type Checking and Duck Typing:** ABCs can be used for type checking with tools like mypy, and they complement duck typing in Python.

**Key Features:**

- **ABCMeta:** A metaclass used to create ABCs.
- **@abstractmethod:** A decorator to mark methods as abstract, forcing subclasses to implement them.
- **register():** A method to register a concrete class as a virtual subclass of an ABC, enhancing flexibility.

**Common Use Cases:**

- Creating interfaces for custom container types (e.g., custom lists, dictionaries).
- Defining base classes for different types of objects in a domain model (e.g., shapes, animals).
- Implementing design patterns that rely on abstract classes (e.g., strategy pattern, template pattern).
- Facilitating testing by mocking abstract classes.
- Incorporating type hints for static type checking.



4. How can we achieve data abstraction?
ans:Data abstraction boils down to focusing on the "what" without worrying about the "how." Here's how it works:

**Imagine you're at a restaurant:**

* **Data is the food:** Ingredients, recipes, cooking methods are all complex details.
* **Data abstraction is the menu:** It tells you what you can order (dishes) without revealing how it's made (recipes, kitchen).

**To achieve data abstraction:**

1. **Define what data matters:** Decide the essential details users need (dish names, ingredients).
2. **Hide the rest:** Keep internal complexities (recipes, cooking) hidden.
3. **Provide clear interfaces:** Offer ways to interact with the data (ordering, eating).

By focusing on the essential and hiding the complexity, you achieve data abstraction. This makes things simpler for users and keeps the system cleaner.

Think of it like a magic box: You put data in, perform operations through buttons, and get results without seeing the internal machinery. That's data abstraction in action!

ans:
Nope, you can't directly create an instance of an abstract class. Think of it like a recipe book - you can't eat the book itself, you need to use the recipes to make the actual food (the concrete objects). Abstract classes serve as templates, defining what things should be like (methods and properties), but they're not finished products themselves.
